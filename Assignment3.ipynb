```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/survey_data.csv" , skipinitialspace=True)
df
```

```python
df1 =  pd.DataFrame(df, columns = ['GENDER', 'PARTICIPATION.TYPE.FOLLOW', 'PARTICIPATION.TYPE.USE.APPLICATIONS', 'PARTICIPATION.TYPE.USE.DEPENDENCIES', 'PARTICIPATION.TYPE.CONTRIBUTE', 'PARTICIPATION.TYPE.OTHER'])
man, woman = ['Man'], ['Woman']
# selecting rows based on condition
df1_man = df1.loc[df1['GENDER'].isin(man)]
df1_women = df1.loc[df1['GENDER'].isin(woman)]
```

```python
df1_man
```

```python
df1_women
```

```python
# men

# follow
cnt_man_follow = df1_man['PARTICIPATION.TYPE.FOLLOW'].value_counts()
ratio_man_follow = cnt_man_follow[1] / 3387
print('follow', ratio_man_follow)
# use applications
cnt_man_use_app = df1_man['PARTICIPATION.TYPE.USE.APPLICATIONS'].value_counts()
ratio_man_use_app = cnt_man_use_app[1] / 3387
print('use applications', ratio_man_use_app)
# use dependencies
cnt_man_use_dep = df1_man['PARTICIPATION.TYPE.USE.DEPENDENCIES'].value_counts()
ratio_man_use_dep = cnt_man_use_dep[1] / 3387
print('use dependencies', ratio_man_use_dep)
# contribute
cnt_man_contrib = df1_man['PARTICIPATION.TYPE.CONTRIBUTE'].value_counts()
ratio_man_contrib = cnt_man_contrib[1] / 3387
print('contribution', ratio_man_contrib)
```

```python
# women

# follow
cnt_woman_follow = df1_women['PARTICIPATION.TYPE.FOLLOW'].value_counts()
ratio_woman_follow = cnt_woman_follow[1] / 125
print('follow', ratio_woman_follow)
# use applications
cnt_woman_use_app = df1_women['PARTICIPATION.TYPE.USE.APPLICATIONS'].value_counts()
ratio_woman_use_app = cnt_woman_use_app[1] / 125
print('use applications', ratio_woman_use_app)
# use dependencies
cnt_woman_use_dep = df1_women['PARTICIPATION.TYPE.USE.DEPENDENCIES'].value_counts()
ratio_woman_use_dep = cnt_woman_use_dep[1] / 125
print('use dependencies', ratio_woman_use_dep)
# contribute
cnt_woman_contrib = df1_women['PARTICIPATION.TYPE.CONTRIBUTE'].value_counts()
ratio_woman_contrib = cnt_woman_contrib[1] / 125
print('contribution', ratio_woman_contrib)
```

```python
import matplotlib.pyplot as plt
X = ['Follow', 'Use applications', 'Use dependencies', 'Contribute']
data = [[ratio_man_follow, ratio_man_use_app, ratio_man_use_dep, ratio_man_contrib],
[ratio_woman_follow, ratio_woman_use_app, ratio_woman_use_dep, ratio_woman_contrib]]
Ymen = data[0]
Zwomen = data[1]

X_axis = np.arange(0, 6, 1.8)
  
plt.bar(X_axis - 0.2, Ymen, 0.4, label = 'Men')
plt.bar(X_axis + 0.2, Zwomen, 0.4, label = 'Women')
  
plt.xticks(X_axis, X)
plt.xlabel("Participation type")
plt.ylabel("Participation rate")
plt.title("Participation for women and men in each activity")
plt.legend()
plt.savefig('participation.png')
plt.show()
```

```python
df3 =  pd.DataFrame(df, columns = ['FORMAL.EDUCATION', 'PROVIDED.HELP'])
df3 = df3.dropna()
df3
```

```python
less_than = df3.loc[(df3['FORMAL.EDUCATION'] == 'Less than secondary (high) school') & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_less_than = less_than / 126
vocation = df3.loc[(df3['FORMAL.EDUCATION'] == 'Vocational/trade program or apprenticeship') & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_vocation = vocation / 127
phd = df3.loc[(df3['FORMAL.EDUCATION'] == 'Doctorate (Ph.D.) or other advanced degree (e.g. M.D., J.D.)') & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_phd = phd / 256
secondary = df3.loc[(df3['FORMAL.EDUCATION'] == 'Secondary (high) school graduate or equivalent') & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_secondary = secondary / 375
college = df3.loc[(df3['FORMAL.EDUCATION'] == 'Some college, no degree') & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_college = college / 637
master = df3.loc[(df3['FORMAL.EDUCATION'] == "Master's degree") & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_master = master / 850
bachelor = df3.loc[(df3['FORMAL.EDUCATION'] == "Bachelor's degree") & (df3['PROVIDED.HELP'] == 'Yes')].count()[1]
ratio_bachelor = bachelor / 1320
```

```python
data = [ratio_less_than, ratio_secondary, ratio_vocation, ratio_college, ratio_bachelor, ratio_master, ratio_phd]
data
```

```python
fig = plt.figure()
labels = ["Less than\nsecondary (high) school", "Secondary (high)\nschool graduate or equivalent",  "Vocational/trade program or apprenticeship", "Some college, no degree", "Bachelor's degree", "Master's degree", "Doctorate (Ph.D.) or other advanced degree (e.g. M.D., J.D.)"]

plt.plot(range(len(data)), data, '--o') # Plotting data
plt.ylim([0.65, 0.85])
plt.legend()
plt.savefig('help_ratio.png')
plt.show()
```
